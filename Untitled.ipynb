```python
class area:
    def __init__(self, iso, name):
        self.iso = iso
        self.name = name
        self.records=[]
        self.recordsAvgd=[]
        self.meta=dict()
        self.daysBefore50=0
    def setDaysBefore50(self):
        self.daysBefore50=0
        for r in self.records:
            if r['total_cases']>=50:
                break
            else:
                self.daysBefore50+=1
    def calcSmoothData(self,WINSIZE):
        self.recordsAvgd=list(self.records)
        WINDOW_HALFSIZE=int(WINSIZE/2)
        fields=list(self.records[0].keys())
        fields.remove('date')
        #for field in fields:
        #    self.recordsAvgd[WINDOW_HALFSIZE][field]=0
        #    for r in range(0, WINSIZE):
        #        self.recordsAvgd[WINDOW_HALFSIZE][field]+= self.records[r][field]      
        #    self.recordsAvgd[WINDOW_HALFSIZE][field]/=WINSIZE
        for r in range(WINDOW_HALFSIZE,len(self.records)-WINDOW_HALFSIZE): #start at halfsize+1 if using old code
            for field in fields:
                #self.recordsAvgd[r][field]=(self.recordsAvgd[r-1][field]
                #                             - self.records[r-WINDOW_HALFSIZE-1][field]/WINSIZE
                #                             + self.records[r+WINDOW_HALFSIZE][field]/WINSIZE)
                self.recordsAvgd[r][field]=statistics.mean([q[field] for q in self.records[r-WINDOW_HALFSIZE:r+WINDOW_HALFSIZE+1]])
```

```python
print('Filling in missing days in OWID data and Trimming all data')

def fillInTheBlanks(recs,dates):
    start=-1
    D=0
    for d in range(0,len(dates)):
        if dates[d]==recs[0]['date']:
            start=d
    if start==-1:
        start=0
        for Dd in range(0,len(recs)):
            if dates[start]==recs[Dd]['date']:
                D=Dd
        if D==0:
            raise Exception('date malformed')
    dates=dates[start:]
    recs=recs[D:]
    
    for d in range(1,len(dates)-1):
        if (recs[d]['date']!=dates[d]):
            temp=recs[d-1]
            temp['new_cases']=0
            temp['new_deaths']=0
            #temp['new_recovered']=0
            temp['date']=dates[d]
            #print("inserting:"+dates[d]+"\tbefore"+recs[d]['date'])
            recs=recs[:d]+[temp]+recs[d:]
    return recs
            
            
def trimData(data,month='02'):
    extra=set()
    under=set()
    extradays=dict()
    for c in data:
        days=-1
        for r in range(0,len(data[c].records)):
            #print(data[c].records[r]['date'])
            if data[c].records[r]['date'].split('-')[1]==month and data[c].records[r]['date'].split('-')[2]=='01':
                days=r
        if days==-1:
            under.add(c)
        else:
            extra.add(c)
            extradays[c]=days
    for c in extra:
        data[c].records=data[c].records[extradays[c]:]
    for c in under:
        underdays=0
        for date in [x['date'] for x in data['IND'].records]:
            if date==data[c].records[0]['date']:
                break
            else:
                underdays+=1
        temp=dict()
        for f in data[c].records[0].keys():
            temp[f]=data[c].records[0][f]
        temp['date']='dummy data-'+str(underdays)
        data[c].records=[temp]*underdays +data[c].records
    for c in data:
        data[c].setDaysBefore50()

dates=[x['date'] for x in dataJH['IND'].records]

for c in dataOWID:
    dataOWID[c].records=fillInTheBlanks(dataOWID[c].records,dates)
    
trimData(dataJH)

    
trimData(dataOWID)
dataIN.records=dataIN.records[2:]
dataIN.setDaysBefore50()

```

```python
peakedJH=set()
notPeakedJH=set()
for c in dataJH:
    active=[r['total_active'] for r in dataJH[c].recordsAvgd[WINDOW_HALFSIZE:-WINDOW_HALFSIZE]]
    max_all=max(active)
    max_recent=max(active[len(active)-10:])
    if max_recent<max_all:
        peakedJH.add(c)
    else:
        notPeakedJH.add(c)
peakedOWID=set()
notPeakedOWID=set()
for c in dataJH:
    active=[r['total_active'] for r in dataJH[c].recordsAvgd[WINDOW_HALFSIZE:-WINDOW_HALFSIZE]]
    max_all=max(active)
    max_recent=max(active[len(active)-10:])
    if max_recent<max_all:
        peakedOWID.add(c)
    else:
        notPeakedOWID.add(c)
peaked=peakedOWID.intersection(peakedJH)
#peaked=set(peakedJH)
for c in peakedOWID.intersection(peakedJH):
#for c in peakedJH:
    if -1 in dataOWID[c].meta.values():
        peaked.remove(c)
    if c in peaked and dataOWID[c].malformed:
        peaked.remove(c)
    if c in peaked and dataOWID[c].meta['population']<50000:
        peaked.remove(c)
    if c in peaked and max([r['total_active'] for r in dataJH[c].records])<2000:
        peaked.remove(c)
print(len(peaked))
max_active=[]
max_active_percapita=[]
total_at_max_active=[]
total_percapita_at_max_active=[]
population=[]
density=[]
median_age=[]
gdp_per_capita=[]
total_tested_at_max_active=[]
total_tested_percapita_at_max_active=[]
total_tested_per_total_cases_at_max_active=[]
total_per_total_tested_at_max_active=[]
for c in peaked:
    max_active.append(max([r['total_active'] for r in dataJH[c].recordsAvgd[WINDOW_HALFSIZE:-WINDOW_HALFSIZE]]))
    max_active_percapita.append(max_active[-1]/dataOWID[c].meta['population'])
    total_at_max_active.append(max([r['total_cases']*(r['total_active']==max_active[-1]) for r in dataJH[c].recordsAvgd[WINDOW_HALFSIZE:-WINDOW_HALFSIZE]]))
    total_percapita_at_max_active.append(total_at_max_active[-1]/dataOWID[c].meta['population'])
    population.append(dataOWID[c].meta['population'])
    density.append(dataOWID[c].meta['density'])
    median_age.append(dataOWID[c].meta['median_age'])
    gdp_per_capita.append(dataOWID[c].meta['gdp_per_capita'])
    
    day_max=(max([r*(dataJH[c].records[r]['total_active']==max_active[-1]) for r in range(0,len(dataJH[c].recordsAvgd[WINDOW_HALFSIZE:-WINDOW_HALFSIZE]))]))
    #print(day_max,len(dataJH[c].records),len(dataOWID[c].records))
    total_tested_at_max_active.append(dataOWID[c].recordsAvgd[WINDOW_HALFSIZE:-WINDOW_HALFSIZE][day_max]['total_tested'])
    total_tested_percapita_at_max_active.append(total_tested_at_max_active[-1]/dataOWID[c].meta['population'])
    total_tested_per_total_cases_at_max_active.append(total_tested_at_max_active[-1]/total_at_max_active[-1])

x=total_percapita_at_max_active
print("##Corelation##")
print("max_active & population",scipy.stats.kendalltau(max_active, population))    
print("total_percapita_at_max_active & population",scipy.stats.kendalltau(x, population)) 
print("total_percapita_at_max_active & density",scipy.stats.kendalltau(x, density)) 
print("total_percapita_at_max_active & median_age",scipy.stats.kendalltau(x, median_age)) 
print("total_percapita_at_max_active & gdp_per_capita",scipy.stats.kendalltau(x, gdp_per_capita)) 
print("total_percapita_at_max_active & total_tested_at_max_active",scipy.stats.kendalltau(x, total_tested_at_max_active)) 
print("total_percapita_at_max_active & total_tested_percapita_at_max_active",scipy.stats.kendalltau(x, total_tested_percapita_at_max_active)) 
print("total_percapita_at_max_active & total_tested_per_total_cases_at_max_active",scipy.stats.kendalltau(x, total_tested_per_total_cases_at_max_active)) 
# Pearson's r  pearsonr
#scipy.stats.spearmanr(x, y)   # Spearman's rho
#scipy.stats.kendalltau(x, y)  # Kendall's tau
   
    
    
poly = PolynomialFeatures(degree=2)
print("##R2 score of Prediction by Polimomial regression (degree=2)##")
Y=max_active_percapita
X_poly = poly.fit_transform([list(r) for r in zip(density)]) 
model = LinearRegression().fit(X_poly, Y)
Y_predic=model.predict(X_poly)
print("max_active  per capita  USING Density: ", r2_score(Y,Y_predic))

Y=max_active
X_poly = poly.fit_transform([list(r) for r in zip(population)]) 
model = LinearRegression().fit(X_poly, Y)
Y_predic=model.predict(X_poly)
print("max_active  USING population: ", r2_score(Y,Y_predic))

Y=max_active
X_poly = poly.fit_transform([list(r) for r in zip(population,density,median_age,gdp_per_capita)]) 
model = LinearRegression().fit(X_poly, Y)
Y_predic=model.predict(X_poly)
print("max_active  USING population, median age, gdp_per_c and density: ", r2_score(Y,Y_predic))

Y=max_active_percapita
X_poly = poly.fit_transform([list(r) for r in zip(density,median_age,gdp_per_capita)]) 
model = LinearRegression().fit(X_poly, Y)
Y_predic=model.predict(X_poly)
print("max_active_per_capita USING median age, gdp_per_c and density: ", r2_score(Y,Y_predic))
#print("max_active_per_capita USING median age, gdp_per_c and density: ", model.coef_)

Y=max_active_percapita
X_poly = poly.fit_transform([list(r) for r in zip(median_age)]) 
model = LinearRegression().fit(X_poly, Y)
Y_predic=model.predict(X_poly)
print("max_active_per_capita USING median age: ", r2_score(Y,Y_predic))

Y=max_active_percapita
X_poly = poly.fit_transform([list(r) for r in zip(density)]) 
model = LinearRegression().fit(X_poly, Y)
Y_predic=model.predict(X_poly)
print("max_active_per_capita USING density: ", r2_score(Y,Y_predic))

Y=max_active_percapita
X_poly = poly.fit_transform([list(r) for r in zip(gdp_per_capita)]) 
model = LinearRegression().fit(X_poly, Y)
Y_predic=model.predict(X_poly)
print("max_active_per_capita USING gdp_per_c: ", r2_score(Y,Y_predic))

Y=total_percapita_at_max_active
X_poly = poly.fit_transform([list(r) for r in zip(density,median_age,total_tested_percapita_at_max_active)]) 
model = LinearRegression().fit(X_poly, Y)
Y_predic=model.predict(X_poly)
print("total_percapita_at_max_active USING median age, total_tested_percapita_at_max_active and density: ", r2_score(Y,Y_predic))
#print("max_active_per_capita USING median age, gdp_per_c and density: ", model.coef_)



print([list(r) for r in zip(density,median_age,gdp_per_capita)][-1])
print(X_poly[-1])

print(poly.fit_transform([[2,3,5]]))

f, axes = plt.subplots(1, 1, figsize=(10, 10), sharex=False)
axes.set_yscale("log")
axes.set_xscale("log")
axes.scatter(density,max_active_percapita)
```
